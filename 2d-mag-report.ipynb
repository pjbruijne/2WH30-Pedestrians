```python
import matplotlib.pyplot as plt
import numpy as np

from ising.checkerboard import simulate
import ising.data_farm as farm
from ising.data_farm import magnetization_from_lattices, magnetization_convergence
from ising.lattices import square_random_lattice as random_lattice
```

```python
lattice = random_lattice(50, 50)
```

```python
magnetization_convergence(magnetization_from_lattices(simulate(lattice.copy(), 1, 2.3, 20)), 1, 2.3)
plt.savefig("report-figures/2d/2dconvergence.pdf", bbox_inches="tight")
```

```python
h_J = 0
farm.magnetization_plot(h_J, (0, 1.1))
Ts = np.linspace(0.1, 15, 1000)
from ising.analytic_results import magnetization2d
plt.plot(Ts, list(map(magnetization2d, Ts)))
plt.legend(["numerical", "analytical"])
plt.savefig(f"simulation-results/2d comparison.pdf")
```

```python
import numpy as np

es = np.linspace(0, 10, 1000)
boltzmann1 = lambda e: np.exp(-e/1)
boltzmann23 = lambda e: (1/2.3)*np.exp(-e/2.3)
boltzmann5 = lambda e: (1/5)*np.exp(-e/5)

plt.plot(es, list(map(boltzmann1, es)))
plt.plot(es, list(map(boltzmann23, es)))
plt.plot(es, list(map(boltzmann5, es)))
plt.legend([r"$1/\beta = 1$",r"$1/\beta =2.3$",r"$1\beta = 5$"])
plt.xlabel("$H$")
plt.ylabel("$P(H)$")
```
