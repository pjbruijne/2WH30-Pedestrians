```python
from ising.lattices import square_random_lattice
from ising.checkerboard import simulate
from ising.visualize import lattices_to_images, upscale_lattices, display_experiment
import matplotlib.pyplot as plt
import numpy as np
```

```python
lattice = square_random_lattice(200, 200)
```

```python
magnetization = np.empty(1000)
i = 0
temps = np.linspace(1, 10, 1000)
for temp in temps:
    magnetization[i] = np.average(
        np.average(simulate(lattice.copy(), h_J=1, T=temp, steps=20), axis=(0, 1))[-10:]
    )
    i += 1
```

```python
plt.plot(temps, magnetization)
```

```python
lattices = simulate(lattice.copy(), h_J=1, T=4, steps=100)
```

```python
plt.plot(range(101), np.average(lattices, axis=(2, 1)))
```

```python
gif = lattices_to_images(upscale_lattices(lattices, 4))
```

```python
display_experiment(gif)
```
