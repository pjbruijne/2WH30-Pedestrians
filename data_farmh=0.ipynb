```python
import ising.data_farm as farm
from ising import checkerboard
import matplotlib.pyplot as plt
import numpy as np
from ising.lattices import square_random_lattice
```

```python
t_low = np.linspace(0.1, 2, 50)
t_transition = np.linspace(2, 3, 100)
t_high = np.linspace(3, 15, 50)
```

```python
h_J = 0
```

```python
lattice = square_random_lattice(50, 50)
```

Before farming you need to check if the initial state is not problematic!

```python
farm.magnetization_convergence(farm.magnetization_from_lattices(checkerboard.simulate(lattice.copy(), 0, 0.1, 1000)), 0, 0.1)
```

```python
farm.magnetization_convergence(farm.magnetization_from_lattices(checkerboard.simulate(lattice.copy(), 0, 2.3, 40_000)), 0, 2.3)
```

```python
farm.constant_harvester(lattice=lattice.copy(), Ts=t_low, h_J=h_J, iterations=10_000, file_id="test")
farm.constant_harvester(lattice=lattice.copy(), Ts=t_transition, h_J=h_J, iterations=100_000, file_id="test")
farm.constant_harvester(lattice=lattice.copy(), Ts=t_high, h_J=h_J, iterations=10_000, file_id="test")
```
