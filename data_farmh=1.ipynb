```python
import ising.data_farm as farm
from ising import checkerboard
import matplotlib.pyplot as plt
import numpy as np
from ising.lattices import square_random_lattice
```

```python
lattice = square_random_lattice(200, 200)
```

```python
t_low = np.linspace(0.1, 2, 100)
t_transition = np.linspace(2, 3, 100)
t_high = np.linspace(3, 15, 100)
```

Before farming you need to check if the initial state is not problematic!

```python
h_J = 1
```

```python
farm.magnetization_convergence(farm.magnetization_from_lattices(checkerboard.simulate(lattice.copy(), h_J, 2.3, 30)), h_J, 2.3)
```

```python
farm.constant_harvester(lattice=lattice, Ts=t_low, h_J=h_J, iterations=50, file_id="test")
farm.constant_harvester(lattice=lattice, Ts=t_transition, h_J=h_J, iterations=50, file_id="test")
farm.constant_harvester(lattice=lattice, Ts=t_high, h_J=h_J, iterations=50, file_id="test")
```
