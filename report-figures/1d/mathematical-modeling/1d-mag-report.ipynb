```python
import matplotlib.pyplot as plt
import numpy as np

from ising.one_dimensional.checkerboard import simulate
from ising.data_farm import magnetization_from_line_lattices, magnetization_convergence
from ising.one_dimensional.lattice import random_lattice
```

```python
EXTENSION = "png"
```

```python
line_lattice = random_lattice(10000)
```

```python
magnetization_convergence(magnetization_from_line_lattices(simulate(line_lattice.copy(), 1, 1, 10)), 1, 1)
plt.savefig(f"report-figures/1d/1dconvergence.{EXTENSION}", bbox_inches="tight")
```

```python
def magnetization_plot(line_lattice, iterations, h_J, points=1000):

    Ts = np.linspace(0.1, 100, points)
    m = np.empty( (points,) )

    for i, T in enumerate(Ts):
        m[i] = np.average(magnetization_from_line_lattices(simulate(line_lattice.copy(), h_J, T, iterations))[-5:])

    import ising.plot_styling
    

    plt.ylim( (0, 1.1) )

    plt.plot(Ts, abs(m))

    plt.ylabel("Magnetization")
    plt.xlabel("Temperature")
    plt.title(f"$h/J = {h_J}$")
```

```python
h_J = 1
magnetization_plot(line_lattice=line_lattice, iterations=20, h_J=h_J)
plt.savefig(f"report-figures/1d/1d-mag-h={h_J}.{EXTENSION}", bbox_inches="tight")
from ising.analytic_results import magnetization1d
Ts = np.linspace(0.1, 100, 1000)
plt.plot(Ts, magnetization1d(Ts, h_J))

plt.legend(["numerical", "analytical"])
plt.savefig(f"report-figures/1d/1d-mag-h={h_J}comp.{EXTENSION}", bbox_inches="tight")
```
