```python
import matplotlib.pyplot as plt
import numpy as np

from ising.checkerboard import simulate as simulate_checkerboard
from ising.metropolis import simulate as simulate_metropolis

from ising.data_farm import magnetization_from_lattices, magnetization_convergence
from ising.lattices import square_random_lattice as random_lattice

from ising.visualize import lattices_to_images, upscale_lattices, display_experiment, save_gif
```

```python
lattice = random_lattice(50, 50)
h_J = 1
T = 2.3
```

```python
metro = simulate_metropolis(lattice=lattice.copy(), h_J=h_J, T=T, steps=1_000_000)
mag_metro = magnetization_from_lattices(metro)
```

```python
magnetization_convergence(mag_metro, h_J, T)
plt.savefig("report-figures/2d/conv-metropolis.pdf", bbox_inches="tight")
```

```python
gif_metro = lattices_to_images(upscale_lattices(metro[::100], 4))
display_experiment(gif_metro)
```

```python
check = simulate_checkerboard(lattice=lattice.copy(), h_J=h_J, T=T, steps=1_000_000)
mag_check = magnetization_from_lattices(check)
```

```python
magnetization_convergence(mag_check, h_J, T)
plt.savefig("report-figures/2d/conv-checkerboard.pdf", bbox_inches="tight")
```

```python
gif_check = lattices_to_images(upscale_lattices(check, 4))
display_experiment(gif_check)
```
